```python
collection = {}
for i in range(len(data)):
    doc_id = data.loc[i]['product_id']
    title = str(data.loc[i]['product_title'])
    text = ' ' if type(data.loc[i]['product_description']) == float else str(data.loc[i]['product_description'])
    bullet_points = ' ' if type(data.loc[i]['product_bullet_point']) == float else str(data.loc[i]['product_bullet_point'])
    product_brand = ' ' if type(data.loc[i]['product_brand']) == float else str(data.loc[i]['product_brand'])
    color = ' ' if type(data.loc[i]['product_color_name']) == float else str(data.loc[i]['product_color_name'])
    locale = data.loc[i]['product_locale']
    content = '<title> '
    if str(title) != 'nan':
        content += str(title)
    content += ' </title><text> '
    if str(text) != 'nan':
        content += str(text)
    content += ' </text><bullet_points> '
    if str(bullet_points) != 'nan':
        content += str(bullet_points)
    content += ' </bullet_points><brand> '
    if str(product_brand) != 'nan':
        content += str(product_brand)
    content += ' </brand><color> '
    if str(color) != 'nan':
        content += str(color)
    content += ' </color><locale> '
    content += str(locale)
    content += ' </locale>'
    collection[doc_id] = {'docid':doc_id, 'title':title, 'text':text, 'bullet_points':bullet_points, 'brand':product_brand, 'color':color, 'locale':locale,'contents': content}

with open('ESCI-product-collection.jsonl','w') as w:
    for doc_id in collection:
        w.write("{}\n".format(json.dumps(collection[doc_id])))
        
len(data)
```
